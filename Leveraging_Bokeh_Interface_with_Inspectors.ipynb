```python
import pandas as pd
appl_df=pd.read_csv('AAPL.csv')
fb_df=pd.read_csv('FB.csv')
google_df=pd.read_csv('GOOGL.csv')
```

```python
map_stocks_df={'AAPL': appl_df, 'FB': fb_df, 'GOOGL': google_df}
for key, value in map_stocks_df.items():
  value['Date']=value[['Date']].apply(lambda x:pd.to_datetime(x))
```

```python
import bokeh.io
from bokeh.models import ColumnDataSource
appl_df_src = ColumnDataSource(appl_df)
```

```python
from bokeh.plotting import figure, show, output_notebook
output_notebook()
plot=figure(
    x_axis_type='datetime', plot_width=800, plot_height=450,
    title='Apple Closing Price Jan-Sept 2020'
)
plot.circle('Date', 'Close', source=appl_df_src)
show(plot)
```

```python
from bokeh.models import HoverTool
hover=HoverTool(tooltips=None, mode='vline')
```

```python
plot=figure(
    x_axis_type='datetime', plot_width=800, plot_height=450,
    title='Apple Closing Price Jan-Sept 2020'
)
plot.line('Date', 'Close', source=appl_df_src)
plot.circle('Date', 'Close', source=appl_df_src, fill_color='lightgrey',
            alpha=0.1, size=8, hover_fill_color='firebrick', hover_alpha=0.5)
plot.add_tools(hover)
show(plot)
```

```python
hover_new=HoverTool(
    tooltips=[('Date', '@Date{%F}'), ('Volume', '@Volume{0.00 a}'), ('Close', '$@Close{0.2f}')], 
    formatters={'@Date':'datetime'},
    mode='mouse'
    )
```

```python
plot=figure(
    x_axis_type='datetime', plot_width=800, plot_height=450,
    title='Apple Closing Price Jan-Sept 2020'
)
plot.line('Date', 'Close', source=appl_df_src)
plot.circle('Date', 'Close', source=appl_df_src, fill_color='lightgrey',
            alpha=0.1, size=8, hover_fill_color='firebrick', hover_alpha=0.5)
plot.add_tools(hover_new)
show(plot)
```
