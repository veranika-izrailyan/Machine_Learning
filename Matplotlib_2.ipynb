```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Bar Charts**

```python
df=pd.read_csv('fifa20_players_stats.csv', index_col=0, usecols=[x for x in range(74)])
df.head()
```

```python
df.shape
```

```python
count_countries=df[['Country', 'Name']].groupby('Country').count().rename(columns={'Name':'Number of Players'})
```

```python
count_countries_sorted=count_countries.sort_values(by='Number of Players', ascending=False)
```

```python
top_50=count_countries_sorted.head(50)
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.bar(top_50.index, top_50['Number of Players'], color='green')
ax.set_ylabel('Number of Players for each Country')
ax.set_xticklabels(top_50.index, rotation=90)
plt.show()
```
