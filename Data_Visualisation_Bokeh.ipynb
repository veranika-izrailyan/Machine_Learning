```python
import pandas as pd
appl_df=pd.read_csv('AAPL.csv')
fb_df=pd.read_csv('FB.csv')
google_df=pd.read_csv('GOOGL.csv')
```

```python
fb_df.info()
```

```python
map_stocks_df={'AAPL': appl_df, 'FB': fb_df, 'GOOGL': google_df}
for key, value in map_stocks_df.items():
  value['Date']=value[['Date']].apply(lambda x:pd.to_datetime(x))
```

```python
fb_df.info()
```

```python
import bokeh.io
from bokeh.plotting import figure, show, output_notebook
```

```python
output_notebook()
```

```python
plot=figure(x_axis_type='datetime')
plot.line(fb_df['Date'].to_list(), fb_df['Close'].to_list())
show(plot)
```

```python
from bokeh.models import ColumnDataSource
fb_df_src = ColumnDataSource(fb_df)
```

```python
fb_df_src.data['Close'][0:10]
```

```python
plot=figure(x_axis_type='datetime', 
            plot_width=800, 
            plot_height=450, 
            title='FB Closing Price',
            tools="pan, wheel_zoom, box_zoom, reset")
plot.line('Date','Close', source=fb_df_src)
plot.title.text_font_size='15pt'
plot.title.text_color='orange'
show(plot)
```

```python
from bokeh.models import WheelZoomTool, WheelPanTool, BoxZoomTool
```

```python
plot=figure(x_axis_type='datetime', 
            plot_width=800, 
            plot_height=450, 
            title='FB Closing Price',
            x_axis_label='Date',
            y_axis_label='Closing Price',
            # tools=[WheelZoomTool(), WheelPanTool(), BoxZoomTool()]
            toolbar_location=None,
            tools=''
            )
plot.line('Date','Close', source=fb_df_src, color='orange', width=3, alpha=0.8, legend_label='FB')
plot.circle('Date','Close',source=fb_df_src, fill_color='white', size=3)
plot.title.text_font_size='15pt'
plot.title.text_color='orange'
# plot.axis.axis_label_text_font_size='12pt'
plot.xaxis.axis_label_text_font_size='15pt'
plot.yaxis.axis_label_text_font_size='12pt'
show(plot)
```

```python
from bokeh.palettes import Colorblind3
```

```python
new_plot1 = figure(
    plot_width=800, plot_height=450,
    x_axis_type='datetime',
    title='Closing Price Jan-Sept 2020'
)
for name, data, color in zip(map_stocks_df.keys(), map_stocks_df.values(), Colorblind3):
  data['Date']=data[['Date']].apply(lambda x: pd.to_datetime(x))
  new_plot1.line(data['Date'], data['Close'], line_width=3, color=color, legend_label=name)

new_plot1.legend.location='top_left'
show(new_plot1)
```

```python
def daily_change(row):
  return row.pct_change()
```

```python
for key, value in map_stocks_df.items():
  value['Returns']=value[['Close']].apply(lambda x:daily_change(x))
```

```python
new_plot2 = figure(
    plot_width=800, plot_height=450,
    x_axis_type='datetime',
    title='Closing Price Jan-Sept 2020'
)
for name, data, color in zip(map_stocks_df.keys(), map_stocks_df.values(), Colorblind3):
  data['Date']=data[['Date']].apply(lambda x: pd.to_datetime(x))
  new_plot2.line(
      data['Date'], data['Returns'], line_width=1, color=color, legend_label=name,
      muted_color=color, muted_alpha=0.2
      )

new_plot2.legend.location='top_left'
new_plot2.legend.background_fill_color='lightblue'
new_plot2.legend.click_policy='mute'
show(new_plot2)
```
