```python
import pandas as pd
appl_df=pd.read_csv('AAPL.csv')
fb_df=pd.read_csv('FB.csv')
google_df=pd.read_csv('GOOGL.csv')
```

```python
map_stocks_df={'AAPL': appl_df, 'FB': fb_df, 'GOOGL': google_df}
for key, value in map_stocks_df.items():
  value['Date']=value[['Date']].apply(lambda x:pd.to_datetime(x))
```

```python
def daily_change(row):
  return row.pct_change()
```

```python
for name, data in zip(map_stocks_df.keys(), map_stocks_df.values()):
  data['returns']=data[['Close']].apply(lambda x: daily_change(x))
```

```python
import bokeh
from bokeh.plotting import figure, show, output_notebook
```

```python
output_notebook()
```

```python
def creating_figure(src_df, title, plot_width=1000, plot_height=700):
  p=figure(
      plot_width=plot_width, plot_height=plot_height,
      x_axis_type='datetime', title=title,
      tools='pan, wheel_zoom, box_zoom, reset'
  )
  p.title.text_font_size='12pt'
  p.line('Date', 'Close', source=src_df, line_width=3, color='blue')
  return p
```

```python
from bokeh.models import ColumnDataSource
```

```python
apple_src_df=ColumnDataSource(appl_df)
fb_src_df=ColumnDataSource(fb_df)
google_src_df=ColumnDataSource(google_df)
```

```python
from bokeh.layouts import row
```

```python
p1=creating_figure(
    apple_src_df, title='Apple Closing Price Jan-Sept 2020',
    plot_width=450,
    plot_height=300
)
```

```python
p2=creating_figure(
    fb_src_df, title='FB Closing Price Jan-Sept 2020',
    plot_width=450,
    plot_height=300
)
```

```python
p3=creating_figure(
    google_src_df, title='Google Closing Price Jan-Sept 2020',
    plot_width=450,
    plot_height=300
)
```

```python
row_plots=row([p1,p2,p3])
show(row_plots)
```

```python
from bokeh.layouts import column
```

```python
column_plots=column([p1,p2,p3])
show(column_plots)
```

```python
p4=creating_figure(
    apple_src_df, title='Apple Closing Price Jan-Sept 2020',
    plot_width=750,
    plot_height=300
)
```

```python
def compare_returns(src_df1, src_df2, title, plot_width=1000, plot_height=700):
  returns1=src_df1.data['returns']
  returns2=src_df2.data['returns']

  p=figure(
      plot_width=plot_width, plot_height=plot_height, title=title,
      tools='pan, wheel_zoom, box_zoom, reset'
  )
  p.title.text_font_size='12pt'
  p.circle(returns1, returns2, color='blue')
  return p
```

```python
p12=compare_returns(
    apple_src_df, fb_src_df, title='Returns Dependency between Apple and FB',
    plot_width=450, plot_height=300
)
p13=compare_returns(
    google_src_df, fb_src_df, title='Returns Dependency between Apple and Google',
    plot_width=450, plot_height=300
)
```

```python
row_plots2=row([p12,p13], sizing_mode='scale_width')
layout=column([p4, row_plots2], sizing_mode='scale_width')
show(layout)
```

```python
def creating_figure_returns(src_df, title, plot_width=1000, plot_height=700):
  p=figure(
      plot_width=plot_width, plot_height=plot_height,
      x_axis_type='datetime', title=title,
      tools='pan,wheel_zoom,box_zoom, reset'
  )
  p.title.text_font_size='12pt'
  p.circle('Date','returns',source=src_df, line_width=3,color='blue')
  return p
```

```python
p5 = creating_figure_returns(apple_src_df, title='Apple Returns Jan-Sept 2020', plot_width=450, plot_height=300)
p6 = creating_figure_returns(fb_src_df, title='FB Returns Jan-Sept 2020', plot_width=450, plot_height=300)
p7 = creating_figure_returns(google_src_df, title='Google Returns Jan-Sept 2020', plot_width=450, plot_height=300)
```

```python
from bokeh.layouts import gridplot
layout=gridplot([[p1,p5],[p2,p6],[p3,p7]])
show(layout)
```

```python
from bokeh.models import Panel, Tabs
```

```python
tab1=Panel(child=row([p1,p5]), title='Apple')
tab2=Panel(child=row([p2,p6]), title='FB')
tab3=Panel(child=row([p3,p7]), title='Google')
tab_obj=Tabs(tabs=[tab1, tab2, tab3])
show(tab_obj)
```
